Tentuka akar-akar real dari persamaan:
$$
f(x) = -0.5x^2 + 2.5x + 4.5
$$
- Secara grafis
- Dengan menggunakan persamaan kuadrat (rumus abc)
- Dengan menggunakan metode biseksi. Gunakan tebakan awal $x_l = 5$ dan $x_u = 10$

```python
def func_01(x):
    return -0.5*x**2 + 2.5*x + 4.5
```

```python
a = -0.5
b = 2.5
c = 4.5
D = b**2 - 4*a*c
x1_true = (-b - math.sqrt(D))/(2*a)
x2_true = (-b + math.sqrt(D))/(2*a)
print("x1_true = %18.10f" % (x1_true) )
print("x2_true = %18.10f" % (x2_true) )
```

```python
x = np.linspace(-2.0,8.0,500)
y = func_01(x)
plt.clf()
plt.plot(x, y)
plt.grid(alpha=0.3)
```

```python
x1, err_rel = bisection(func_01, 6.0, 8.0, verbose=True, Niter=3)
```

```python
print("err_rel  = %18.10e" % err_rel)
print("err_true = %18.10e" % abs(x1 - x1_true))
```
