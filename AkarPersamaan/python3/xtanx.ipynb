A problematic function:
$$
f(x) = x - \tan(x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("svg")
%matplotlib inline
```

```python
def f(x):
    return x - np.tan(x)
```

```python
x = np.linspace(17.0, 18.0, 1000)
y = f(x)
plt.clf()
plt.plot(x, y, marker="o")
```

Near the singularities at $\pi/2, 3\pi/2, 5\pi/2, \ldots$

```python
plt.clf()
fig, axs = plt.subplots(4,1,figsize=(10,20))
for i in range(0,4):
    cc = 2*i+1
    x = np.linspace(cc*np.pi/2 - 0.01, cc*np.pi/2 + 0.01, 1000)
    y = f(x)
    axs[i].plot(x, y, marker="o")
```
