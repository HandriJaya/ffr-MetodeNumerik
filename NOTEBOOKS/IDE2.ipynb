Mari kita mulai dengan plot sederhana

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*pi,100)
y = np.sin(x)
plt.clf()
plt.plot(x,y)
```

```python
plt.clf()
plt.plot(x,y,marker="o")
```

```python
plt.clf()
plt.plot(x,y,marker="s")
```

```python
plt.clf()
plt.plot(x,y,marker="^")
```

```python
plt.clf()
plt.plot(x,y,marker="s",label="first plot")
plt.plot(x,np.cos(2*x),marker="o", label="second plot")
plt.grid()
plt.legend()
```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, 10))
ax.set_ylim((-2, 10))

line, = ax.plot(np.arange(6), np.random.rand(6), lw=2)
```
