Plot beberapa fungsi, tampilkan juga label untuk sumbu x, y, judul, dan legenda.

```python
plt.clf()
x = np.linspace(0, 2, 100)

plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label="cubic")

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")

plt.legend()
```

```python
plt.clf()
x = np.linspace(0, 2, 100)
plt.plot(x, x, "r--", label="linear")
plt.plot( x, x**2, "b-", label="quadratic")
plt.plot(x, x**3, "g^", label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")
plt.legend()
```

```python
plt.clf()
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, marker="o", label="sin")
plt.plot(x, y2, marker="*", label="cos")
plt.plot(x, y1*y2, marker="v", label="sin*cos")
plt.legend()
plt.grid()
plt.xlabel
plt.title("Plot Urang Euy")
```

Mari tulis persamaan lagi:
$$
\int f(x)\,\mathrm{d}x = F(x) + C

Gelombang kotak:

$$
f(x) = \frac{4}{\pi}
\sum_{n=1,3,5,...}^{\infty}
\frac{1}{n}
\sin\left(

$$


```python
L = 4.0
x = np.linspace(0,L,100)
suku1 = np.sin(2*np.pi*x/L)/1
suku2 = np.sin(2*3*np.pi*x/L)/3
suku3 = np.sin(2*5*np.pi*x/L)/5
suku4 = np.sin(2*7*np.pi*x/L)/7

plt.clf()
plt.plot(x, 4/np.pi*(suku1 + suku2 + suku3 + suku4))
plt.title(r"$f(x) = \frac{4}{\pi}\sum_{n=1,3,5,...}\frac{1}{n}\sin\left(\frac{n\pi x}{L}\right)$")
```

```python
plt.clf()
x = np.linspace(0,np.pi,100)
fig, ax_lst = plt.subplots(2, 2)
ax_lst[0][0].plot(np.cos(x))
ax_lst[0][1].plot(np.sin(x))
ax_lst[1][0].plot(np.sin(x)*np.cos(x))
ax_lst[1][1].plot(np.random.rand(20))
```

```python
plt.clf()
fig, ax = plt.subplots(2, 3)
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.clf()

plt.subplot(211)
plt.plot(t1, f(t1), "bo", t2, f(t2), "k")

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
```

```python
names = ["data_1", "data_2", "data_3"]
values = [1, 15, 45]

plt.clf()

f, (ax1,ax2,ax3) = plt.subplots(1,3, sharex="col", sharey="row",figsize=(10, 3))

ax1.bar(names, values)
ax2.scatter(names, values)
ax3.plot(names, values)

plt.suptitle("Categorical Plotting")
```

## Stream plot

```python
x = np.linspace(0,1,4)
y = np.linspace(0,1,3)
print("x = ", x)
print("y = ", y)

X, Y = np.meshgrid(x,y)
print("X = ", X)
print("Y = ", Y)
```

```python
x = np.linspace(0,5,10)
y = np.linspace(0,5,10)
X, Y = np.meshgrid(x,y)
U = 2*X
V = -5*Y

plt.clf()
plt.streamplot(X,Y,U,V)
plt.title("$V = 2x\^i - 5y\^j$")
```

```python
x = np.arange(0,5,0.025)
y = np.arange(0,5,0.025)
X, Y = np.meshgrid(x,y)
Z = X*Y

plt.clf()
cs = plt.contour(X,Y,Z,20)
plt.clabel(cs, inline=1, fontsize=8)
plt.title("Z = X*Y")
```
