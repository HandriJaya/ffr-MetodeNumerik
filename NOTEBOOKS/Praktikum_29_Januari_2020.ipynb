```python
Npoints = 100
```

```python
x = np.linspace(-np.pi, np.pi, Npoints)
```

```python
y = np.zeros(Npoints)
y2 = np.zeros(Npoints)
```

```python
for i in range(Npoints):
    y[i] = math.sin(x[i])
    y2[i] = math.cos(x[i])
```

```python
plt.clf()
plt.plot(x, y, label="sin")
plt.plot(x, y2, label="cos")
plt.grid()
plt.legend()
```

```python
Npoints = 200
x = np.linspace(-2*np.pi, np.pi, Npoints)
plt.clf()
plt.plot(x, np.sin(10*x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.sin(x)*np.cos(x), label="sin*cos")
plt.grid()
plt.legend()
plt.ylim(-1.5, 1.5)
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'go', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(131)
plt.plot(t1, f(t1), 'go', t2, f(t2), 'k')

plt.subplot(132)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

plt.subplot(133)
plt.plot(t2, np.sin(2*np.pi*t2), 'b--')
plt.show()
```

```python
names = ['data_1', 'data_2', 'data_3']
values = [1, 15, 45]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
```
