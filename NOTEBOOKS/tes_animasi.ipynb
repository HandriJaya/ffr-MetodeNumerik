```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from matplotlib import animation, rc
```

```python
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()
ax.set_xlim((0.0, 2.0))
ax.set_ylim((-2.0, 2.0))
line, = ax.plot([], [], linewidth=2, marker="o", color="red");
```

```python
def init():
    line.set_data([], [])
    return (line,)
```

```python
def animate(i):
    x = np.linspace(0.0, 2.0, 200)
    y = np.sin(2*np.pi*(x - 0.01*i)) + np.sin(3*2*np.pi*(x - 0.01*i))
    line.set_data(x, y)
    return (line,)
```

```python
fig.set_size_inches(6*1.5, 4*1.5)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=True)
```

```python
HTML( anim.to_html5_video() )
```
